#Normalização

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler().fit(X_train)
X_train = scaler.transform(X_train)
X_test = scaler.transform(X_test)
```

#Machine Learning


```python
from sklearn.svm import SVC

def trainSVM(X_train, y_train, seed):
  model = SVC(random_state=seed)
  model.fit(X_train, y_train)
  return model

model = trainSVM(X_train, y_train, seed)
```

#Previsão e avaliação do modelo de treinamento aplicado a base de teste

```python
from sklearn.metrics._plot.confusion_matrix import ConfusionMatrixDisplay
def predict_and_evaluate(X_test, y_test):

  y_pred = model.predict(X_test)

  from sklearn.metrics import accuracy_score
  accuracy = accuracy_score(y_test, y_pred)
  print('Acurácia: ', accuracy)

  from sklearn.metrics import cohen_kappa_score
  kappa = cohen_kappa_score(y_test, y_pred)
  print('Kappa: ', kappa)

  from sklearn.metrics import f1_score
  f1 = f1_score(y_test, y_pred)
  print('F1: ', f1)

  from sklearn.metrics import confusion_matrix
  confMatrix = confusion_matrix(y_test, y_pred)

  ax = plt.subplot()
  sns.heatmap(confMatrix, annot = True, fmt = ".0f")
  plt.xlabel('Previsto')
  plt.ylabel('Real')
  plt.title('Matriz de Confusão')

  ax.xaxis.set_ticklabels(['Não Pagou', 'Pagou'])
  ax.yaxis.set_ticklabels(['Não Pagou', 'Pagou'])
  plt.show()

predict_and_evaluate(X_test, y_test)
```

#Otimização com hiperparâmetros

```python
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import classification_report

tuned_parameters = [{'kernel':['rbf'], 'gamma':[1e-3, 1e-4], 'C':[1, 10, 100, 1000]}]

print("# Otimizando hiperparâmetros para F1 score")
print()

model = GridSearchCV(SVC(), tuned_parameters, scoring='f1')
model.fit(X_train, y_train)

y_pred = model.predict(X_test)
```
